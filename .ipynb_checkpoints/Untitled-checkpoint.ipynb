```python
import numpy as np
import cv2
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from tqdm import tqdm
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import argparse
import pathlib
from collections import Counter

```

```python

class ImageProcessor:
    def __init__(self, target_size=(64, 64)):
        self.target_size = target_size
        
    def load_and_preprocess_image(self, image_path):
        img = cv2.imread(image_path)
        if img is None:
            raise ValueError(f"Could not load image at {image_path}")
            
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img_resized = cv2.resize(img_gray, self.target_size)
        img_normalized = img_resized / 255.0
        
        return img_normalized.flatten()

    def load_dataset(self, base_path):
        features = []
        labels = []
        
        parent_dir = os.path.dirname(base_path)
        
        stress_path = base_path
        print(f"Loading stress images from: {stress_path}")
        for img_name in tqdm(os.listdir(stress_path), desc="Loading stress images"):
            if img_name.endswith(('.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG')):
                img_path = os.path.join(stress_path, img_name)
                try:
                    img_features = self.load_and_preprocess_image(img_path)
                    features.append(img_features)
                    labels.append(1)
                except Exception as e:
                    print(f"Error processing {img_path}: {str(e)}")
        
        no_stress_path = os.path.join(parent_dir, 'NoStress')
        print(f"Loading non-stress images from: {no_stress_path}")
        for img_name in tqdm(os.listdir(no_stress_path), desc="Loading non-stress images"):
            if img_name.endswith(('.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG')):
                img_path = os.path.join(no_stress_path, img_name)
                try:
                    img_features = self.load_and_preprocess_image(img_path)
                    features.append(img_features)
                    labels.append(0)
                except Exception as e:
                    print(f"Error processing {img_path}: {str(e)}")
        
        return np.array(features), np.array(labels)
```

```python
processor = ImageProcessor(target_size=(64, 64))
stress_dir = '/home/axyl/codes/College/ml/KDEF/KDEF/Train/Stress'
features, labels = processor.load_dataset(stress_dir)
```

```python
class LogisticRegression:
    def __init__(self, learning_rate=0.001, num_iterations=10000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None
        self.losses = []
        
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-np.clip(z, -250, 250)))
    
    def fit(self, X, y, batch_size=32):
        n_samples, n_features = X.shape
        
        # Initialize parameters
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        # Mini-batch gradient descent
        for iteration in tqdm(range(self.num_iterations), desc="Training"):
            # Shuffle data
            indices = np.random.permutation(n_samples)
            
            # Mini-batch training
            for i in range(0, n_samples, batch_size):
                batch_indices = indices[i:min(i + batch_size, n_samples)]
                X_batch = X[batch_indices]
                y_batch = y[batch_indices]
                
                # Forward pass
                linear_pred = np.dot(X_batch, self.weights) + self.bias
                predictions = self.sigmoid(linear_pred)
                
                # Calculate gradients
                dw = (1/len(batch_indices)) * np.dot(X_batch.T, (predictions - y_batch))
                db = (1/len(batch_indices)) * np.sum(predictions - y_batch)
                
                # Update parameters
                self.weights -= self.learning_rate * dw
                self.bias -= self.learning_rate * db
                
            # Calculate loss for the entire epoch
            predictions = self.sigmoid(np.dot(X, self.weights) + self.bias)
            loss = -np.mean(y * np.log(predictions + 1e-15) + (1 - y) * np.log(1 - predictions + 1e-15))
            self.losses.append(loss)
    
    def predict(self, X):
        linear_pred = np.dot(X, self.weights) + self.bias
        y_pred = self.sigmoid(linear_pred)
        return (y_pred >= 0.5).astype(int)

def evaluate_model(y_true, y_pred):
    accuracy = np.mean(y_true == y_pred)
    
    tn = np.sum((y_true == 0) & (y_pred == 0))
    fp = np.sum((y_true == 0) & (y_pred == 1))
    fn = np.sum((y_true == 1) & (y_pred == 0))
    tp = np.sum((y_true == 1) & (y_pred == 1))
    
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1_score,
        'confusion_matrix': {
            'tn': tn, 'fp': fp,
            'fn': fn, 'tp': tp
        }
    }

def plot_training_curve(losses):
    plt.figure(figsize=(10, 6))
    plt.plot(losses)
    plt.title('Training Loss Over Time')
    plt.xlabel('Iteration')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.show()

processor = ImageProcessor(target_size=(64, 64))

# Load and process dataset
print("Loading and processing images...")

# Print dataset statistics
print(f"\nDataset statistics:")
print(f"Total samples: {len(labels)}")
print(f"Stress samples: {np.sum(labels == 1)}")
print(f"No-stress samples: {np.sum(labels == 0)}")

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.2, random_state=42
)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train model
print("\nTraining logistic regression...")
model = LogisticRegression(learning_rate=0.005, num_iterations=1500)
model.fit(X_train_scaled, y_train)

# Plot training curve
plot_training_curve(model.losses)

# Evaluate model
y_pred = model.predict(X_test_scaled)
metrics = evaluate_model(y_test, y_pred)

print("\nModel Performance Metrics:")
print(f"Accuracy: {metrics['accuracy']:.4f}")
print(f"Precision: {metrics['precision']:.4f}")
print(f"Recall: {metrics['recall']:.4f}")
print(f"F1 Score: {metrics['f1_score']:.4f}")

print("\nConfusion Matrix:")
print(f"True Negative: {metrics['confusion_matrix']['tn']}")
print(f"False Positive: {metrics['confusion_matrix']['fp']}")
print(f"False Negative: {metrics['confusion_matrix']['fn']}")
print(f"True Positive: {metrics['confusion_matrix']['tp']}")

```

```python
class KNNClassifier:
    def __init__(self, k=5):
        self.k = k
        self.X_train = None
        self.y_train = None
    
    def fit(self, X, y):
        self.X_train = X
        self.y_train = y
    
    def predict(self, X):
        predictions = []
        
        for sample in tqdm(X, desc="KNN Prediction"):
            distances = np.sqrt(np.sum((self.X_train - sample)**2, axis=1))
            k_indices = np.argsort(distances)[:self.k]
            k_nearest_labels = self.y_train[k_indices]
            most_common = Counter(k_nearest_labels).most_common(1)
            predictions.append(most_common[0][0])
        
        return np.array(predictions)
```

```python
def compare_models(X_train, X_test, y_train, y_test, learning_rate=0.001, num_iterations=1000, k=5):
    logreg = LogisticRegression(learning_rate=learning_rate, num_iterations=num_iterations)
    knn = KNNClassifier(k=k)
    print("\nTraining Logistic Regression...")
    logreg.fit(X_train, y_train)
    logreg_pred = logreg.predict(X_test)
    logreg_metrics = evaluate_model(y_test, logreg_pred)
    print("\nTraining KNN...")
    knn.fit(X_train, y_train)
    knn_pred = knn.predict(X_test)
    knn_metrics = evaluate_model(y_test, knn_pred)
    
    plot_roc_curves(logreg, knn, X_test, y_test)
    
    return logreg_metrics, knn_metrics

def plot_roc_curves(logreg, knn, X_test, y_test):
    logreg_probs = logreg.sigmoid(np.dot(X_test, logreg.weights) + logreg.bias)
    knn_probs = []
    for sample in X_test:
        distances = np.sqrt(np.sum((knn.X_train - sample)**2, axis=1))
        k_indices = np.argsort(distances)[:knn.k]
        k_nearest_labels = knn.y_train[k_indices]
        prob = np.mean(k_nearest_labels)
        knn_probs.append(prob)
    knn_probs = np.array(knn_probs)
    
    thresholds = np.linspace(0, 1, 100)
    logreg_tpr, logreg_fpr = [], []
    knn_tpr, knn_fpr = [], []
    
    for threshold in thresholds:
        logreg_pred = (logreg_probs >= threshold).astype(int)
        tn = np.sum((y_test == 0) & (logreg_pred == 0))
        fp = np.sum((y_test == 0) & (logreg_pred == 1))
        fn = np.sum((y_test == 1) & (logreg_pred == 0))
        tp = np.sum((y_test == 1) & (logreg_pred == 1))
        
        logreg_tpr.append(tp / (tp + fn) if (tp + fn) > 0 else 0)
        logreg_fpr.append(fp / (fp + tn) if (fp + tn) > 0 else 0)
        
        knn_pred = (knn_probs >= threshold).astype(int)
        tn = np.sum((y_test == 0) & (knn_pred == 0))
        fp = np.sum((y_test == 0) & (knn_pred == 1))
        fn = np.sum((y_test == 1) & (knn_pred == 0))
        tp = np.sum((y_test == 1) & (knn_pred == 1))
        
        knn_tpr.append(tp / (tp + fn) if (tp + fn) > 0 else 0)
        knn_fpr.append(fp / (fp + tn) if (fp + tn) > 0 else 0)
    
    # Plot ROC curves
    plt.figure(figsize=(10, 6))
    plt.plot(logreg_fpr, logreg_tpr, label='Logistic Regression')
    plt.plot(knn_fpr, knn_tpr, label='KNN')
    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves Comparison')
    plt.legend()
    plt.grid(True)

def main():
    print(f"\nDataset statistics:")
    print(f"Total samples: {len(labels)}")
    print(f"Stress samples: {np.sum(labels == 1)}")
    print(f"No-stress samples: {np.sum(labels == 0)}")
    
    X_train, X_test, y_train, y_test = train_test_split(
        features, labels, test_size=0.2, random_state=42
    )
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    logreg_metrics, knn_metrics = compare_models(
        X_train_scaled, 
        X_test_scaled, 
        y_train, 
        y_test,
        learning_rate=0.01,
        num_iterations=100,
        k=5
    )
    
    print("\nLogistic Regression Performance:")
    print(f"Accuracy: {logreg_metrics['accuracy']:.4f}")
    print(f"Precision: {logreg_metrics['precision']:.4f}")
    print(f"Recall: {logreg_metrics['recall']:.4f}")
    print(f"F1 Score: {logreg_metrics['f1_score']:.4f}")
    print("\nConfusion Matrix:")
    print(f"True Negative: {logreg_metrics['confusion_matrix']['tn']}")
    print(f"False Positive: {logreg_metrics['confusion_matrix']['fp']}")
    print(f"False Negative: {logreg_metrics['confusion_matrix']['fn']}")
    print(f"True Positive: {logreg_metrics['confusion_matrix']['tp']}")
    
    print("\nKNN Performance:")
    print(f"Accuracy: {knn_metrics['accuracy']:.4f}")
    print(f"Precision: {knn_metrics['precision']:.4f}")
    print(f"Recall: {knn_metrics['recall']:.4f}")
    print(f"F1 Score: {knn_metrics['f1_score']:.4f}")
    print("\nConfusion Matrix:")
    print(f"True Negative: {knn_metrics['confusion_matrix']['tn']}")
    print(f"False Positive: {knn_metrics['confusion_matrix']['fp']}")
    print(f"False Negative: {knn_metrics['confusion_matrix']['fn']}")
    print(f"True Positive: {knn_metrics['confusion_matrix']['tp']}")

if __name__ == "__main__":
    main()
```

```python
class DecisionTreeClassifier:
    def __init__(self, max_depth=5, min_samples_split=2):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.tree = None
        
    class Node:
        def __init__(self, feature=None, threshold=None, left=None, right=None, value=None):
            self.feature = feature
            self.threshold = threshold
            self.left = left
            self.right = right
            self.value = value
    
    def _entropy(self, y):
        classes, counts = np.unique(y, return_counts=True)
        probabilities = counts / len(y)
        return -np.sum(probabilities * np.log2(probabilities + 1e-10))
    
    def _information_gain(self, parent, left_child, right_child):
        p = len(left_child) / len(parent)
        return self._entropy(parent) - p * self._entropy(left_child) - (1-p) * self._entropy(right_child)
    
    def _best_split(self, X, y):
        best_gain = -1
        best_feature = None
        best_threshold = None
        
        n_samples, n_features = X.shape
        
        for feature in range(n_features):
            thresholds = np.unique(X[:, feature])
            
            for threshold in thresholds:
                left_mask = X[:, feature] <= threshold
                right_mask = ~left_mask
                
                if np.sum(left_mask) < self.min_samples_split or np.sum(right_mask) < self.min_samples_split:
                    continue
                
                gain = self._information_gain(y, y[left_mask], y[right_mask])
                
                if gain > best_gain:
                    best_gain = gain
                    best_feature = feature
                    best_threshold = threshold
        
        return best_feature, best_threshold
    
    def _build_tree(self, X, y, depth=0):
        n_samples, n_features = X.shape
        n_classes = len(np.unique(y))
        
        if (depth >= self.max_depth or 
            n_samples < self.min_samples_split or 
            n_classes == 1):
            leaf_value = np.argmax(np.bincount(y))
            return self.Node(value=leaf_value)
        
        best_feature, best_threshold = self._best_split(X, y)
        
        if best_feature is None:
            leaf_value = np.argmax(np.bincount(y))
            return self.Node(value=leaf_value)
        
        left_mask = X[:, best_feature] <= best_threshold
        right_mask = ~left_mask
        
        left_tree = self._build_tree(X[left_mask], y[left_mask], depth + 1)
        right_tree = self._build_tree(X[right_mask], y[right_mask], depth + 1)
        
        return self.Node(best_feature, best_threshold, left_tree, right_tree)
    
    def fit(self, X, y):
        self.tree = self._build_tree(X, y)
    
    def _predict_single(self, x, node):
        if node.value is not None:
            return node.value
        
        if x[node.feature] <= node.threshold:
            return self._predict_single(x, node.left)
        return self._predict_single(x, node.right)
    
    def predict(self, X):
        predictions = []
        for x in tqdm(X, desc="Decision Tree Prediction"):
            pred = self._predict_single(x, self.tree)
            predictions.append(pred)
        return np.array(predictions)

def compare_all_models(X_train, X_test, y_train, y_test, learning_rate=0.01, num_iterations=100, k=5, max_depth=5):
    logreg = LogisticRegression(learning_rate=learning_rate, num_iterations=num_iterations)
    knn = KNNClassifier(k=k)
    dtree = DecisionTreeClassifier(max_depth=max_depth)
    
    print("\nTraining Logistic Regression...")
    logreg.fit(X_train, y_train)
    logreg_pred = logreg.predict(X_test)
    logreg_metrics = evaluate_model(y_test, logreg_pred)
    
    print("\nTraining KNN...")
    knn.fit(X_train, y_train)
    knn_pred = knn.predict(X_test)
    knn_metrics = evaluate_model(y_test, knn_pred)
    
    print("\nTraining Decision Tree...")
    dtree.fit(X_train, y_train)
    dtree_pred = dtree.predict(X_test)
    dtree_metrics = evaluate_model(y_test, dtree_pred)
    
    plot_all_roc_curves(logreg, knn, dtree, X_test, y_test)
    
    return logreg_metrics, knn_metrics, dtree_metrics

def plot_all_roc_curves(logreg, knn, dtree, X_test, y_test):
    plt.figure(figsize=(10, 6))
    
    logreg_probs = logreg.sigmoid(np.dot(X_test, logreg.weights) + logreg.bias)
    plot_roc_curve(y_test, logreg_probs, 'Logistic Regression')
    
    knn_probs = []
    for sample in X_test:
        distances = np.sqrt(np.sum((knn.X_train - sample)**2, axis=1))
        k_indices = np.argsort(distances)[:knn.k]
        k_nearest_labels = knn.y_train[k_indices]
        prob = np.mean(k_nearest_labels)
        knn_probs.append(prob)
    plot_roc_curve(y_test, np.array(knn_probs), 'KNN')
    
    dtree_probs = []
    for x in X_test:
        pred = dtree._predict_single(x, dtree.tree)
        dtree_probs.append(pred)
    plot_roc_curve(y_test, np.array(dtree_probs), 'Decision Tree')
    
    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves Comparison')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_roc_curve(y_true, y_prob, model_name):
    thresholds = np.linspace(0, 1, 100)
    tpr, fpr = [], []
    
    for threshold in thresholds:
        y_pred = (y_prob >= threshold).astype(int)
        tn = np.sum((y_true == 0) & (y_pred == 0))
        fp = np.sum((y_true == 0) & (y_pred == 1))
        fn = np.sum((y_true == 1) & (y_pred == 0))
        tp = np.sum((y_true == 1) & (y_pred == 1))
        
        tpr.append(tp / (tp + fn) if (tp + fn) > 0 else 0)
        fpr.append(fp / (fp + tn) if (fp + tn) > 0 else 0)
    
    plt.plot(fpr, tpr, label=model_name)

def main():
    print(f"\nDataset statistics:")
    print(f"Total samples: {len(labels)}")
    print(f"Stress samples: {np.sum(labels == 1)}")
    print(f"No-stress samples: {np.sum(labels == 0)}")
    
    X_train, X_test, y_train, y_test = train_test_split(
        features, labels, test_size=0.2, random_state=42
    )
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    logreg_metrics, knn_metrics, dtree_metrics = compare_all_models(
        X_train_scaled, 
        X_test_scaled, 
        y_train, 
        y_test,
        learning_rate=0.01,
        num_iterations=100,
        k=5,
        max_depth=5
    )
    print("\nLogistic Regression Performance:")
    print(f"Accuracy: {logreg_metrics['accuracy']:.4f}")
    print(f"Precision: {logreg_metrics['precision']:.4f}")
    print(f"Recall: {logreg_metrics['recall']:.4f}")
    print(f"F1 Score: {logreg_metrics['f1_score']:.4f}")
    
    print("\nKNN Performance:")
    print(f"Accuracy: {knn_metrics['accuracy']:.4f}")
    print(f"Precision: {knn_metrics['precision']:.4f}")
    print(f"Recall: {knn_metrics['recall']:.4f}")
    print(f"F1 Score: {knn_metrics['f1_score']:.4f}")
    
    print("\nDecision Tree Performance:")
    print(f"Accuracy: {dtree_metrics['accuracy']:.4f}")
    print(f"Precision: {dtree_metrics['precision']:.4f}")
    print(f"Recall: {dtree_metrics['recall']:.4f}")
    print(f"F1 Score: {dtree_metrics['f1_score']:.4f}")

if __name__ == "__main__":
    main()
```
